```python
import numpy as np
import pandas as pd

nomi = ["alice","bob","carlo","daniele","ernesto","francesco"]
città = ["napoli", "roma", "milano", "torino", "firenze", "palermo"]

data = []

for i in range(50):

    data.append([np.random.choice(nomi),np.random.randint(0,99),np.random.choice(città),np.random.randint(1000, 3000)])

df = pd.DataFrame(data, columns = ["Nome","Età","Città","Salario"])

df
```

```python
for elemento in df.columns:
    df[elemento].fillna(df[elemento].mean, inplace = True)
```

```python
def cat_età(numero):

    if numero < 18:
        return "Giovane"
    elif numero < 66:
        return "Adulto"
    else:
        return "Senior"

df["Categoria Età"]=df["Età"].apply(cat_età)

df
```
