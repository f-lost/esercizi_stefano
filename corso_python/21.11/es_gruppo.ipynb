```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('imdb.csv')

print("\nInformazioni Data Frame IMDB\n\n")

display(df.info())
```

```python
df = df.drop(["backdrop_path","poster_path", "homepage", "tconst","overview", "tagline"], axis = 1)

df.isnull().sum()
```

```python
for column in df:
    print(column, df[column].nunique() ,"elementi unici")
```

```python
df["adult"].replace("False", 0, inplace = True)
df["adult"].replace("True", 1, inplace = True)
```

```python
plt.figure(figsize=(20,8))
sns.violinplot(data = df, x = "averageRating", y= "runtime", hue="adult")
plt.title('ViolinPlot IMDB')
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="kde", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()

print("\n\n--------------------\n\n")

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="scatter", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()

print("\n\n--------------------\n\n")

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="reg", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()

print("\n\n--------------------\n\n")

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="hist", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```
